### Test the LinearRegression class

```python
from linear_regression import LinearRegression
```

```python
import numpy as np 
rng = np.random.default_rng(seed=42)

m = 200
X = 2 * rng.random((m, 1))
y = 4 + 3 * X + rng.standard_normal((m, 1))
```

```python
# visualize the created dataset
import matplotlib.pyplot as plt 

plt.scatter(X[:, 0], y)
plt.xlabel("X1")
plt.ylabel("y")

plt.show()
```

```python
# train our linear regression model on the dataset
lin_reg_model = LinearRegression()

lin_reg_model.fit(X, y)
y_hat = lin_reg_model.predict(X)
```

```python
print(f""" The model parameters:
           weight: {np.squeeze(lin_reg_model.coef_)}
           bias: {np.squeeze(lin_reg_model.intercept_)} 
      """)
```

```python
# visualize the predicted values compared to the actual values
plt.scatter(X[:, 0], y)
plt.plot(X[:, 0], y_hat, color="r", label="Predictions")
plt.xlabel("X1")
plt.ylabel("y")

plt.legend()
plt.show()
```
